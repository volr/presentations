```python
import tensorflow as tf
import numpy as np
```

```python
with tf.Session() as sess:
  print(sess.list_devices())
```

```python
mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

x_train = np.array([x.reshape(784) for x in x_train])
x_test = np.array([x.reshape(784) for x in x_test])
```

```python
def train_model():
    model = tf.keras.models.Sequential([
      tf.keras.layers.Dense(784, activation=tf.nn.relu, input_shape=(784,)),
      tf.keras.layers.Dense(10, activation=tf.nn.softmax)
    ])
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    model.fit(x_train, y_train, epochs=5)
    model.evaluate(x_test, y_test)
    return np.array(model.get_weights())
```

```python
for i in range(100):
    weights = train_model()
    np.save('mnist_weights{}.npy'.format(i), weights)
```
